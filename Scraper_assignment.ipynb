# Combining all json files (orginal links + missed ones)

```python
import json

def combine_json_files(file_paths, output_path):
    combined_data = {}

    for file_path in file_paths:
        with open(file_path, 'r') as f:
            data = json.load(f)

        for key, value in data.items():
            if key not in combined_data:
                combined_data[key] = value

    with open(output_path, 'w') as output_file:
        json.dump(combined_data, output_file, indent=4)

    print(f"Combined data saved to {output_path}")


file_paths = ['datasets_output.json', 'datasets2_output.json', 'datasets3_output.json']
output_path = 'combined_datasets_output.json'
combine_json_files(file_paths, output_path)
```

# Creating folders and keeping metadata and its respective csv files in it

```python
import json
import os
import requests

def process_json_data(input_json, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    with open(input_json, 'r') as f:
        data = json.load(f)

    for idx, (title, content) in enumerate(data.items()):
        print(f"Processing {idx + 1}/{len(data)}: {title}")
        folder_path = os.path.join(output_folder, title)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)

        metadata_file_path = os.path.join(folder_path, 'metadata.txt')
        with open(metadata_file_path, 'w') as metadata_file:
            for line in content.get('metadata', []):
                metadata_file.write(line + "\n")

        datasets = content.get('datasets', {})
        for dataset_name, dataset_url in datasets.items():
            try:
                print(f"Downloading {dataset_name}...")

                response = requests.get(dataset_url)
                if response.status_code == 200:
                    csv_file_path = os.path.join(folder_path, dataset_name + ".csv")
                    with open(csv_file_path, 'wb') as csv_file:
                        csv_file.write(response.content)
                    print(f"Downloaded: {dataset_name}")
                else:
                    print(f"Skipping {dataset_name}: URL not reachable")
            except Exception as e:
                print(f"Error downloading {dataset_name}: {e}")

    print("Processing complete!")


input_json = 'combined_datasets_output.json'
output_folder = 'Telangana datasets'
process_json_data(input_json, output_folder)
```
